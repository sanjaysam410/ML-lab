```python
import random

# -------------------------------
# Decode chromosome (binary → int)
# -------------------------------
def decode(chromosome):
    """
    Converts a binary chromosome (list of 0s/1s)
    into its decimal integer value.
    Example: [1,0,1] -> 5
    """
    return int("".join(map(str, chromosome)), 2)


# -------------------------------
# Fitness function (x^2)
# -------------------------------
def fitness(chromosome):
    """
    Fitness = square of the decoded chromosome value.
    """
    x = decode(chromosome)
    return x * x


# -------------------------------
# Main Genetic Algorithm
# -------------------------------
def genetic_algorithm(chromosome_length, population_size, generations):
    """
    Simple Genetic Algorithm that evolves a population
    to maximize the function f(x) = x^2.
    """

    # Step 1: Initialize random population
    population = [
        [random.randint(0, 1) for _ in range(chromosome_length)]
        for _ in range(population_size)
    ]

    # Step 2: Evolve for a fixed number of generations
    for gen in range(generations):

        # --- Evaluate fitness of current population
        fitness_values = [fitness(ch) for ch in population]

        # Best chromosome in this generation
        best_index = fitness_values.index(max(fitness_values))
        best_chromosome = population[best_index]

        print(
            f"Generation {gen+1}: "
            f"Best = {best_chromosome}, "
            f"x = {decode(best_chromosome)}, "
            f"fitness = {fitness(best_chromosome)}"
        )

        # --- Create new population
        new_population = []

        while len(new_population) < population_size:
            # Select two random parents
            parent1 = random.choice(population)
            parent2 = random.choice(population)

            # Single-point crossover
            crossover_point = random.randint(1, chromosome_length - 1)
            child = parent1[:crossover_point] + parent2[crossover_point:]

            # Mutation (flip each bit with 10% chance)
            for i in range(chromosome_length):
                if random.random() < 0.1:
                    child[i] = 1 - child[i]

            new_population.append(child)

        # --- Elitism: keep the best from old population
        worst_index = min(range(population_size), key=lambda i: fitness(new_population[i]))
        new_population[worst_index] = best_chromosome.copy()

        # Update population
        population = new_population

    # Step 3: Final best solution
    final_best = max(population, key=fitness)
    print("\nFinal Best Chromosome:", final_best)
    print("Final x:", decode(final_best))
    print("Final Fitness (x^2):", fitness(final_best))


# -------------------------------
# Run the GA
# -------------------------------
genetic_algorithm(chromosome_length=5, population_size=6, generations=5)
```
