```python
import pandas as pd
from sklearn.metrics import accuracy_score,precision_score
import matplotlib.pyplot as plt
```

```python
mp=pd.read_csv("Mobileprice.csv")
```

```python
from sklearn.model_selection import train_test_split
```

```python
mp_train,mp_test=train_test_split(mp,test_size=0.2)
```

```python
mp_train_x=mp_train.iloc[:,0:-1]
mp_train_y=mp_train.iloc[:,-1]
mp_test_x=mp_test.iloc[:,0:-1]
mp_test_y=mp_test.iloc[:,-1]
```

```python
from sklearn.ensemble import RandomForestClassifier
random_forest=RandomForestClassifier(n_estimators=100,random_state=67)
random_forest.fit(mp_train_x,mp_train_y)
random_forest_pred=random_forest.predict(mp_test_x)
random_forest_acc=accuracy_score(mp_test_y,random_forest_pred)
random_forest_acc
```

```python
random_forest_prec=precision_score(mp_test_y,random_forest_pred,average='macro')
random_forest_prec
```

```python
from sklearn.tree import DecisionTreeClassifier
dec_tree=DecisionTreeClassifier(criterion='entropy')
dec_tree.fit(mp_train_x,mp_train_y)
```

```python
pred_dec_tree=dec_tree.predict(mp_test_x,)
dec_tree_precision_score=precision_score(mp_test_y,pred_dec_tree,average='macro')
dec_tree_precision_score
dec_tree_accuracy_score=accuracy_score(mp_test_y,pred_dec_tree)
dec_tree_accuracy_score
```

```python
model=['DT','RF']
acc=[dec_tree_accuracy_score,random_forest_acc]
prec=[dec_tree_precision_score,random_forest_acc]
```

```python
plt.bar(model,acc,color=('red','yellow'))
plt.xlabel('Classifier')
plt.ylabel('Accuracy')
plt.title('Accuracy comparision between DecisionTree and RandomForest')
plt.grid(axis='y')
```

```python
plt.bar(model,acc,color=('yellow','black'))
plt.xlabel('Classifier')
plt.ylabel('Precision')
plt.title('Precision comparision between DecisionTree and RandomForest')
plt.grid(axis='y')
```
