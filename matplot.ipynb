```python
import matplotlib.pyplot as plt
```

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]
plt.figure(figsize=(5,5))
plt.plot(x,y,marker="*",color="red")
plt.title("simple")
plt.xlabel('H')
plt.ylabel('J')
```

```python
x=["A","B","C","D"]
y=[10,30,40,20]
c=["red",'yellow','blue','green']
plt.bar(x,y,color=c,width=0.2,)
```

```python
x=["A","B","C","D"]
y=[10,30,40,20]
c=["red",'yellow','blue','green']
plt.barh(x,y,color=c,height=0.2)
```

```python
std_perf=["Excellent","Good","Average","Poor"]
std_values=[15,25,12,8]
plt.pie(std_values,labels=std_perf,colors=["Black","brown","blue","yellow"],autopct="%2.2f%%",explode=[0.2,0,0,0.1],shadow=True)
```
